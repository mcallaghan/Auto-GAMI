```python
# Filter out only the included studies

df = df[df['Include']==True]
print(df['APSIS ID'].unique().shape)
df.head()
```

```python
# Make a table to show which documents were coded by whom
df['coded'] = 1
df['CI'] = df['0.1 Coder initials'].str.strip()#.str[0].str.lower() # < uncomment if we want to group together coders with the same first initial
pair_lookup = df[['APSIS ID','CI','coded']].pivot_table(index="APSIS ID",columns="CI").fillna(0)

pair_lookup
```

```python
# Get the combinations of pairs and count how many documents they coded together

pairs = []
combinations = []
for name, group in df[['APSIS ID','CI']].groupby("APSIS ID")['CI']:#.aggregate(lambda x: ', '.join(x))#.to_frame()
    if len(set([x.strip() for x in group])) > 1:   
        for p in itertools.combinations(group,2):
            if set(p) in combinations:
                continue
            if p[0]==p[1]:
                continue
            combinations.append(set(p))
            pp = pair_lookup[(pair_lookup[('coded',p[0])]==1) & (pair_lookup[('coded',p[1])]==1)]
            n = pp.shape[0]
            if n > 0:
                x = {"p0": p[0], "p1": p[1], "n": n}
                pairs.append(x)
len(pairs)
    
#pair_lookup
```

```python
# How many documents have each pair coded?
pair_df = pd.DataFrame.from_dict(pairs)
pair_df.sort_values('n',ascending=False)
```

```python
# Load the master data

master_df = pd.read_csv('data/GAMI_Database_with_bibinfo.csv')
print(master_df.shape)
master_df = master_df.merge(df[~df.duplicated(["APSIS ID","content"])][["APSIS ID", "content"]],how="left")

columns = ["APSIS ID","Title","content"]
y_column = "1.2.Sector"

# This is a function than turns a column with multiple values separated by "|||" into separate binary columns

def widen_column(df, column, sep='|||'):
    columns = set([])
    for x in df[column].unique():
        for s in str(x).split(sep):
            s = s.strip()
            if s =="nan" or s=="None" or s=="" or s is None or "no information" in s.lower() or s=="Other":
                continue
            columns.add(s.strip())
            
    for c in columns:
        print(repr(c))
        df[c] = 0
        df.loc[df[column].str.contains(c)==True,c] = 1
        
    return df, columns

# Remove some typos
master_df['GEOG_REGION'] = master_df['GEOG_REGION'].str.replace('IslandsEurope', 'Islands|Europe')
master_df['GEOG_REGION'] = master_df['GEOG_REGION'].str.replace('AmericaAfrica', 'America|Africa')
master_df['1.2.Sector'] = master_df['1.2.Sector'].str.replace('Health Health', 'Health')
            
# Run the widen column function on sectors, topics, actors, etc.
print("SECTORS\n")
master_df, sectors = widen_column(master_df, '1.2.Sector')

print("\nTOPICS\n")
master_df, topics = widen_column(master_df, '1.3.Cross.cutting.topics')

print("\nACTORS\n")
master_df, actors = widen_column(master_df, '2.1.1.Actors.Institutions', '|')

print("\nREGIONS\n")
master_df, regions = widen_column(master_df, 'GEOG_REGION', '|')
    
print("\nHAZARDS\n")
master_df, hazards = widen_column(master_df, '3.3.1.Hazards', '|')
    
master_df.head()


master_df = master_df[(pd.notna(master_df['APSIS ID']))].reset_index(drop=True)
master_df
```

```python
# y is the variable (in this case a matrix of 1s and 0s as wide as the number of sectors and as long as the )
master_df['labels'] = list(master_df[sectors].values)
y = np.matrix(master_df[sectors])
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier


pipeline = Pipeline([
    ('vect', TfidfVectorizer()),
    ('clf', OneVsRestClassifier(SVC(probability=True))),
])

parameters = [
    {
        'vect__max_df': (0.75,),
        'vect__min_df': (2, 5),
        'vect__ngram_range': ((1, 2),),  
        'clf__estimator__kernel': ['rbf'], 
        'clf__estimator__gamma': [1e-3, 1e-4],
        'clf__estimator__C': [100, 1000],
        'clf__estimator__class_weight': ['balanced',]
    },
    {
        'vect__max_df': (0.75,),
        'vect__min_df': (2, 5),
        'vect__ngram_range': ((1, 2),),  
        'clf__estimator__kernel': ['linear'], 
        'clf__estimator__C': [1, 10],
        'clf__estimator__class_weight': ['balanced',]
    }
]

grid_search = GridSearchCV(pipeline, parameters, n_jobs=6, verbose=1, scoring="f1_macro")
```

```python
for i, row in pair_df.sort_values('n',ascending=False).iterrows():    
    ids = pair_lookup[(pair_lookup[('coded',row["p0"])]==1) & (pair_lookup[('coded',row["p1"])]==1)].index
    test = master_df[master_df['APSIS ID'].isin(ids)].index
    train = master_df[~master_df['APSIS ID'].isin(ids)].index
    
    grid_search.fit(master_df.loc[train, 'content'].astype(str), y[train])
    
    y_pred = grid_search.predict_proba(master_df.loc[test,'content'])
    
    preds = pd.DataFrame(y_pred)
    preds.columns = sectors
    preds["APSIS ID"] = list(master_df.loc[test]["APSIS ID"])
    preds["Title"] = list(master_df.loc[test]["Title"])
    
    preds.to_csv(f'results/preds_{row["p0"]}_{row["p1"]}_sector.csv',index=False)
    
    break
    
preds.head()
```

```python
# We can also see how the different model candidates performed
results = pd.DataFrame(grid_search.cv_results_)
results.sort_values('mean_test_score',ascending=False).head()
```

```python
from sklearn.model_selection import KFold

# Now we just split into 5 test sets, and for each of these make a prediction based on the rest of the data
predictions = master_df[['Title','content']]
kf = KFold(n_splits=5, shuffle=True)
for train, test in kf.split(master_df.index):
    grid_search = GridSearchCV(pipeline, parameters, n_jobs=8, verbose=1, scoring="f1_macro")
    grid_search.fit(master_df.loc[train, 'content'].astype(str), y[train])
    print("best macro f1 score: ", grid_search.best_score_)
    y_pred = grid_search.predict_proba(master_df.loc[test,'content'].astype(str))
    for i, t in enumerate(sectors):
        predictions.loc[test,t] = y_pred[:,i]
        
predictions.head()
```
